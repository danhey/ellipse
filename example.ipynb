```python
from ellipse import Ellipse, intersection
import matplotlib.pyplot as plt
import jax.numpy as jnp
```

```python
el1 = Ellipse(5,20,1)
el2 = Ellipse(5,10,0)
```

```python
x, y = intersection(el1, el2)
```

```python
fig, ax = plt.subplots()

el1.plot(ax)
el2.plot(ax)

ax.scatter(x, y)

plt.savefig('example.png')
```

```python
%timeit intersection(el1, el2)
```
